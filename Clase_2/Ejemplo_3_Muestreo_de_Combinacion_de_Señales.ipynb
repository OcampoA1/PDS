# Ejemplo 3. Muestreo de una combinación de señales analógicas.

```python
# Importar las librerias necesarias
import numpy as np
import matplotlib.pyplot as plt
```

```python
f1 = 25     # Frecuencia de la señal 1
f2 = 150    # Frecuencia de la señal 2
f3 = 50     # Frecuencia de la señal 3
fs = 1e3    # Frecuencia de muestreo
t = np.arange(0, 0.1, 1/fs) # Vector de tiempo
signal = (3*np.cos(2*np.pi*f1*t)) + (100*np.sin(2*np.pi*f2*t)) - (np.cos(2*np.pi*f3*t)) # Sumade señales
```

```python
# Grafica de la señal analógica
plt.figure()
plt.plot(t, signal)
plt.xlabel('Tiempo')
plt.ylabel('Amplitud')
plt.grid()
plt.show()
```

La señal anterior esta muestreada correctamente porque se usó una frecuencia de muestro superior a 2 veces la máxima que es 150, o sea, mayor a 300Hz. Si se usa por ejemplo una frecuencia de muestreo de 100Hz, que es inferior a 300Hz que es la tasa de Nyquist mínima en este caso, la señal estaría mal muestreada como se muestra a continuación.

```python
fs = 1e2    # Frecuencia de muestreo nueva
t = np.arange(0, 0.1, 1/fs) # Vector de tiempo
signal = (3*np.cos(2*np.pi*f1*t)) + (100*np.sin(2*np.pi*f2*t)) - (np.cos(2*np.pi*f3*t)) # Suma de señales
```

```python
# Grafica de la señal analógica
plt.figure()
plt.plot(t, signal)
plt.xlabel('Tiempo')
plt.ylabel('Amplitud')
plt.grid()
plt.show()
```

# Ejemplos propios

```python
import numpy as np
import matplotlib.pyplot as plt

f1 = 300
f2 = 100
f3 = 45
fs= 900
t = np.arange(0,0.1,1/fs)
signal = 2*np.sin(2*np.pi*f1*t) + (3/4)*np.cos(2*np.pi*f2*t) - 44*np.sin(2*np.pi*f3*t)
plt.figure()
plt.plot(t, signal)
plt.xlabel('Tiempo')
plt.ylabel('Amplitud')
plt.grid()
plt.show()
```

Se puede observar que si se cumple el criterio de Nyquist, la señal se ve bien muestreada, pues en el ejemplo anterior se puede ver que se cumple el criterio, donde, la frecuencia de muestreo es 900, y dos veces la frecuencia maxima sería 600, es decir, Fs>= 2Fmax. En cambio en el siguiente ejemplo, se toma una frecuencia de muestreo de 70, por tal motivo, la señal no será muestreada de forma optima, no se cumple el criterio de Nyquist

```python
fs= 70
t = np.arange(0,0.1,1/fs)
signal = 2*np.sin(2*np.pi*f1*t) + (3/4)*np.cos(2*np.pi*f2*t) - 44*np.sin(2*np.pi*f3*t)
plt.figure()
plt.plot(t, signal)
plt.xlabel('Tiempo')
plt.ylabel('Amplitud')
plt.grid()
plt.show()
```

La señal queda mal muestreada, debido a que se tomó una frecuencia de muestreo que no cumple el criterio de Nyquist

```python
import numpy as np
import matplotlib.pyplot as plt

f1 = 1000
f2 = 3000
f3 = 400
fs = 6000
t = np.arange(0, 0.01, 1/fs)
signal = 4*np.square(np.sin(2*np.pi*f1*t)) - 4*np.sin(2*np.pi*f2*t) + 6*np.cos(2*np.pi*f3*t)
plt.figure()
plt.plot(t, signal)
plt.xlabel('Tiempo')
plt.ylabel('Amplitud')
plt.grid()
plt.show()
```

Se observa en el ejemplo anterior que el criterio de Nyquist se cumple al borde, es decir, 6000 >= 2*3000, por tal motivo, la señal se empieza a muestrear de manera optima.

```python
fs = 4000
t = np.arange(0, 0.01, 1/fs)
signal = 4*np.square(np.sin(2*np.pi*f1*t)) - 4*np.sin(2*np.pi*f2*t) + 6*np.cos(2*np.pi*f3*t)
plt.figure()
plt.plot(t, signal)
plt.xlabel('Tiempo')
plt.ylabel('Amplitud')
plt.grid()
plt.show()
```

En este caso no se cumple el criterio de Nyquist, pues no se cumple que Fs>=2Fmax, por tal motivo la señal noqueda bien muestreada
