# Ejemplo 2. Como simular ruido blanco a una señal analógica.

```python
# Importar las librerias necesarias
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0, 2*np.pi, 20)     # Vector de tiempo
signal = np.sin(t)                  # Señal
noise = np.random.randn(20)         # ruido blanco
signal_with_noise = signal + noise  # Se añade el ruido a la señal
```

```python
# Grafica de las señales con y sin ruido
plt.figure()
plt.plot(t, signal, 'b', label = 'Señal son ruido')
plt.plot(t, signal_with_noise, 'r', label = 'Señal con ruido')
plt.xlabel('Tiempo')
plt.ylabel('Amplitud')
plt.grid()
plt.legend()
plt.show()
```

# Ejemplos propios

```python
import numpy as np 
import matplotlib.pyplot as plt

#Establecer parámetros
A = 3
F = 10
fs = 100
d  = 1
#parámetros del ruido
minimo = -6
maximo = 6
#señal
t = np.arange(0, d, 1/fs)
s = A*np.cos(2*np.pi*F*t)
ruido = np.random.uniform(low=minimo, high=maximo, size=len(s))
s_con_ruido = s + ruido

plt.plot(t,s, label='Señal original')
plt.plot(t,s_con_ruido, label='Señal con ruido uniforme')
plt.xlabel("Tiempo")
plt.ylabel("Amplitud")
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

#Establecer parámetros
A = 4
F = 5
fs = 1000
d  = 1
#parámetros del ruido
lam = 0.1
#señal
t = np.arange(0, d, 1/fs)
s = A*np.square(np.sin(2*np.pi*F*t))
ruido = np.random.poisson(lam = lam, size=len(s))
s_con_ruido = s + ruido

plt.plot(t,s, label='Señal original')
plt.plot(t,s_con_ruido, label='Señal con ruido Poisson')
plt.xlabel("Tiempo")
plt.ylabel("Amplitud")
plt.grid(True)
plt.legend()
plt.show()
```
