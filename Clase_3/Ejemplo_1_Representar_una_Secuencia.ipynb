# Ejemplo 1. Representar una secuencia.

```python
# Importar las librerias necesarias
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_n = [1, 4, 3, 0, 4, 2, 3, 4, 0, 1, 4] # Secuencia
n = np.arange(len(x_n))                 # Vector de muestras
```

```python
# Grafica de la secuencia
plt.figure()
plt.stem(n, x_n)
plt.xlabel('Numero de muestras')
plt.ylabel('Amplitud')
plt.grid()
plt.show()
```

# Ejemplos propios

```python
import numpy as np
import matplotlib.pyplot as plt

vector = [-2,-3,-4, 0, 0, 0, 9, 7]
t = np.arange(len(vector))
plt.figure()
plt.stem(t, vector)
plt.xlabel('Numero de muestras')
plt.ylabel('Amplitud')
plt.grid()
plt.show()
```

Se representa una secuencia con valores negativos

```python
import numpy as np
import matplotlib.pyplot as plt

vector = [(3/4),-3,4, 0.6, 1, 0, 9, 7]
t = np.arange(len(vector))
plt.figure()
plt.stem(t, vector)
plt.xlabel('Numero de muestras')
plt.ylabel('Amplitud')
plt.grid()
plt.show()
```

Se representa una secuencia con numeros racionales
