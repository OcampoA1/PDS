# Ejemplo2. Generar señales útiles como el delta, escalón, y rampa.

```python
# Importar las librerias necesarias
import numpy as np
import matplotlib.pyplot as plt
```

Es muy importante tener bien definido el vector de muestras, este se genera en base N que es el numero de muestras, se recomienda siempre un N impar.

```python
N = 9
n = np.arange(-N//2+1, N//2+1)
```

#### 1. Función delta

```python
delta = np.zeros(N) # Vector de ceros
delta[N//2] = 1     # Se pone un uno en la mitad

# Grafica de la secuencia
plt.figure()
plt.title('Función delta')
plt.stem(n, delta)
plt.xlabel('Numero de muestras')
plt.ylabel('Amplitud')
plt.grid()
plt.show()
```

#### 2. Función escalón unitario

```python
amplitud_escalon = 1    # Amplitud
desplazamiento = 1      # Desplazamiento del escalón
u = np.heaviside(n-desplazamiento, amplitud_escalon)    # Función escalón

# Otra forma de hacerlo:
# u = np.piecewise(n, n>=1, [1, 0])

# Grafica de la secuencia
plt.figure()
plt.figure()
plt.title('Función escaló|n unitario')
plt.stem(n, u)
plt.xlabel('Numero de muestras')
plt.ylabel('Amplitud')
plt.grid()
plt.show()
```

#### 3. Función rampa

```python
rampa = []              # Vector vacío para rellenar
desplazamiento2 = 2     # Desplazamiento de la rampa
cont = 1                # Contador para generar la rampa
for i in range(N):
    if (i >= (N//2+1)-(-desplazamiento2)):
        rampa.append(cont)
        cont += 1
    else:
        rampa.append(0)

# Grafica de la secuencia
plt.figure()
plt.figure()
plt.title('Función rampa')
plt.stem(n, rampa)
plt.xlabel('Numero de muestras')
plt.ylabel('Amplitud')
plt.grid()
plt.show()
```

# Ejemplos propios

```python
import numpy as np
import matplotlib.pyplot as plt

#Se toma N par
N = 10
n = np.arange(-N//2+1, N//2+1)
delta = np.zeros(N) 
delta[N//2] = 1     

# Grafica de la secuencia
plt.figure()
plt.title('Función delta')
plt.stem(n, delta)
plt.xlabel('Numero de muestras')
plt.ylabel('Amplitud')
plt.grid()
plt.show()
```

Se toma N par, donde se puede observar que en la grafica, en el eje x, de numero de muestras hay un punto más a la izquierda, por tal motivo es que anteriormente se recomienda que sea numero impar.

La función delta inicia en 5, debido a que se puso en alto la señal en la mitad + 4, y como la mitad era 1, entonces se pone en 5 la función delta.

```python
N = 18
n = np.arange(-N//2+1, N//2+1)
u = np.piecewise(n, n>=2, [4, 0])


plt.figure()
plt.figure()
plt.title('Función escalón unitario')
plt.stem(n, u)
plt.xlabel('Numero de muestras')
plt.ylabel('Amplitud')
plt.grid()
plt.show()
```

Función escalón, utilizando np.piecewise, el escalon inicial en 2, tiene amplitud de 4, con 18 muestras.

```python
N = 15
n = np.arange(-N//2+10, N//2+10)
rampa = []              
desplazamiento2 = 3     
cont = 1                
for i in range(N):
    if (i >= (N//2+1)-(-desplazamiento2)):
        rampa.append(cont)
        cont += 1
    else:
        rampa.append(0)

# Grafica de la secuencia
plt.figure()
plt.figure()
plt.title('Función rampa')
plt.stem(n, rampa)
plt.xlabel('Numero de muestras')
plt.ylabel('Amplitud')
plt.grid()
plt.show()
```

La función rampa anterior tiene un numero de muestras igual a 15, un desplazamiento de 3, donde el vector definido, partiendo de este numero de muestras se divide de manera entera por 12, por tal motivo, el escalón comienza en 12.
