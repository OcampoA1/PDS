# Ejemplo 1. Digitalizaci칩n de se침ales anal칩gicas.

```python
# Importar las librerias necesarias
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0, 2*np.pi, 20) # Vector de tiempo
signal = np.sin(t)              # Se침al analogica
```

```python
# Grafica de la se침al anal칩gica y digital
plt.figure()
plt.plot(t, signal, 'b', label = 'Se침al anal칩gica')
plt.stem(t, signal, 'r', label = 'Se침al digitalizada')
plt.xlabel('Tiempo')
plt.ylabel('Amplitud')
plt.grid()
plt.legend()
plt.show()
```

# Ejemplos propios

```python
import numpy as np
import matplotlib.pyplot as plt

#Se establecen los par치metros para la se침al

amplitud = 2
frecuencia = 10
fs = 100
duracion = 1
tiempo = np.arange(0, duracion, 1/fs) #Se establece el vector de tiempo
senal_analogica = amplitud * np.sin(2 * np.pi * frecuencia * tiempo)#Se instancia la se침al anal칩gica
senal_digital = np.round(senal_analogica) #con np.round se aproxima a valores de la se침al analogica a los valores discretos cercanos

plt.subplot(2, 1, 1)
plt.plot(tiempo, senal_analogica)
plt.title('Se침al anal칩gica')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
plt.grid(True)

plt.subplot(2, 1,2)
plt.stem(tiempo, senal_digital)
plt.title('Se침al digitalizada')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
plt.grid()

plt.tight_layout()
plt.show()



```

```python
import numpy as np
import matplotlib.pyplot as plt

#Se establecen los par치metros para la se침al

amplitud = 4
frecuencia = 30
fs = 100
duracion = 1
tiempo = np.arange(0, duracion, 1/fs) #Se establece el vector de tiempo
senal_analogica = amplitud * np.square(np.sin(2 * np.pi * frecuencia * tiempo))#Se igenera se침al cuadrada a partir de una se침al seno
senal_digital = np.round(senal_analogica) #con np.round se aproxima a valores de la se침al analogica a los valores discretos cercanos

plt.subplot(2, 1, 1)
plt.plot(tiempo, senal_analogica)
plt.title('Se침al anal칩gica')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
plt.grid(True)

plt.subplot(2, 1,2)
plt.stem(tiempo, senal_digital)
plt.title('Se침al digitalizada')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
plt.grid()

plt.tight_layout()
plt.show()
```
