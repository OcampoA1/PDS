# Ejemplo 1. Digitalización de señales analógicas.

```python
# Importar las librerias necesarias
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0, 2*np.pi, 20) # Vector de tiempo
signal = np.sin(t)              # Señal analogica
```

```python
# Grafica de la señal analógica y digital
plt.figure()
plt.plot(t, signal, 'b', label = 'Señal analógica')
plt.stem(t, signal, 'r', label = 'Señal digitalizada')
plt.xlabel('Tiempo')
plt.ylabel('Amplitud')
plt.grid()
plt.legend()
plt.show()
```

# Ejemplos propios

```python
import numpy as np
import matplotlib.pyplot as plt

#Se establecen los parámetros para la señal

amplitud = 2
frecuencia = 10
fs = 100
duracion = 1
tiempo = np.arange(0, duracion, 1/fs) #Se establece el vector de tiempo
senal_analogica = amplitud * np.sin(2 * np.pi * frecuencia * tiempo)#Se instancia la señal analógica
senal_digital = np.round(senal_analogica) #con np.round se aproxima a valores de la señal analogica a los valores discretos cercanos

plt.subplot(2, 1, 1)
plt.plot(tiempo, senal_analogica)
plt.title('Señal analógica')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
plt.grid(True)

plt.subplot(2, 1,2)
plt.stem(tiempo, senal_digital)
plt.title('Señal digitalizada')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
plt.grid()

plt.tight_layout()
plt.show()



```

```python
import numpy as np
import matplotlib.pyplot as plt

#Se establecen los parámetros para la señal

amplitud = 4
frecuencia = 30
fs = 100
duracion = 1
tiempo = np.arange(0, duracion, 1/fs) #Se establece el vector de tiempo
senal_analogica = amplitud * np.square(np.sin(2 * np.pi * frecuencia * tiempo))#Se igenera señal cuadrada a partir de una señal seno
senal_digital = np.round(senal_analogica) #con np.round se aproxima a valores de la señal analogica a los valores discretos cercanos

plt.subplot(2, 1, 1)
plt.plot(tiempo, senal_analogica)
plt.title('Señal analógica')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
plt.grid(True)

plt.subplot(2, 1,2)
plt.stem(tiempo, senal_digital)
plt.title('Señal digitalizada')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
plt.grid()

plt.tight_layout()
plt.show()
```
