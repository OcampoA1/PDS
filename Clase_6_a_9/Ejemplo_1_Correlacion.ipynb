# Ejemplo 1. Correlaci√≥n.

```python
# Importar las librerias necesarias
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([1, 4, 2, 5, 1])
y = np.array([4, 2, 5])
Rxy  = np.correlate(x, y, mode='full')
```

```python
# Grafica de la correlaci√≥n
plt.figure()
plt.stem(Rxy)
plt.xlabel('Numero de muestras')
plt.ylabel('Amplitud')
plt.grid()
plt.show()
```

# Ejemplos propios

Se realiza ejemplo de correlaci√≥n

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.array([5, 44, 45, 6, 3])
y = np.array([4, 34, 5, 76, 6])
Rxy  = np.correlate(x, y, mode='full')
plt.figure()
plt.stem(Rxy)
plt.xlabel('Numero de muestras')
plt.ylabel('Amplitud')
plt.grid()
plt.show()
```

Se puede ver la gr√°fica de correlaci√≥n, donde en los puntos mas altos de la se√±al, es donde ambas se√±ales son mas parecidas.

Se realiza ejemplo de autocorrelaci√≥n

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.array([2,4,6,8,10])
y = np.array([2,4,6,8,10])
Rxy  = np.correlate(x, y, mode='full')
plt.figure()
plt.stem(Rxy)
plt.xlabel('Numero de muestras')
plt.ylabel('Amplitud')
plt.grid()
plt.show()
```

En la gr√°fica anterior de autocorrelaci√≥n, se puede ver como en el valor 4 del eje x, est√° en el punto m√°ximo, por tal motivo, eso se puede interpretar como la energ√≠a m√°xima y donde ambas se√±ales, las cuales son iguales, esta en ese momento una encima de la otra.
