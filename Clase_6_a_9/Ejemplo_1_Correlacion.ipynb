# Ejemplo 1. Correlación.

```python
# Importar las librerias necesarias
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([1, 4, 2, 5, 1])
y = np.array([4, 2, 5])
Rxy  = np.correlate(x, y, mode='full')
```

```python
# Grafica de la correlación
plt.figure()
plt.stem(Rxy)
plt.xlabel('Numero de muestras')
plt.ylabel('Amplitud')
plt.grid()
plt.show()
```

# Ejemplos propios

Se realiza ejemplo de correlación

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.array([5, 44, 45, 6, 3])
y = np.array([4, 34, 5, 76, 6])
Rxy  = np.correlate(x, y, mode='full')
plt.figure()
plt.stem(Rxy)
plt.xlabel('Numero de muestras')
plt.ylabel('Amplitud')
plt.grid()
plt.show()
```

Se puede ver la gráfica de correlación, donde en los puntos mas altos de la señal, es donde ambas señales son mas parecidas.

Se realiza ejemplo de autocorrelación

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.array([2,4,6,8,10])
y = np.array([2,4,6,8,10])
Rxy  = np.correlate(x, y, mode='full')
plt.figure()
plt.stem(Rxy)
plt.xlabel('Numero de muestras')
plt.ylabel('Amplitud')
plt.grid()
plt.show()
```

En la gráfica anterior de autocorrelación, se puede ver como en el valor 4 del eje x, está en el punto máximo, por tal motivo, eso se puede interpretar como la energía máxima y donde ambas señales, las cuales son iguales, esta en ese momento una encima de la otra.
